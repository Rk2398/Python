```python
# Performing KMeans clsutering
```

```python
data = pd.read_csv("G:/data sceince/Assignments/Clustering/crime_data.csv")
```

```python
data.head()
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(data.iloc[:,1:])
df_norm
```

```python
# Creating ELBOW curve
```

```python
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,random_state= 0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("ELBOW CURVE")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
```

```python
model = KMeans(n_clusters = 4)
model.fit(df_norm)
model.labels_
```

```python
md = pd.Series(model.labels_)
data['clusterid']=md
```

```python
data
```

```python
data.iloc[:,1:].groupby(data.clusterid).mean()
```

Inference: As compared to rest of the created clusters Places having high murder case rates belongs to 0th cluster, Places having high Assault case rates belongs to 2nd cluster, Places having high rate of urban population belongs to 2nd cluster, similarly places having high rate of rape cases belongs to 2nd cluster
